```python
a = [1,2,3,4,2,3,5,1,2,1,2]
frequency = {}
for char in a:
    if char in frequency:
        frequency[char] += 1
    else:
     frequency[char] = frequency.get(char, 0) + 1

print("Frequency of elements:", frequency)
```

```python

# 82.	Find most frequent element
a = [1,2,3,4,2,3,5,1,2,1,2]
frequent = {}
for i in a:
    if i in frequent:
        frequent[i] += 1
    else:
        frequent[i] = 1

max_count = max(frequent.values())

for key, value in frequent.items():
    if value == max_count:
        print("Most frequent element:", key)
        print("Frequency:", value)

```

```python
a =  [1,2,3,4,2,3,5,1,2,1,2]
max_count = 0
most_frequent = None
for i in a:
    count = a.count(i)
    if count > max_count:
        max_count = count
        most_frequent = i

print("Most frequent element:", most_frequent)
print("Frequency:", max_count)
```

```python
# 83.	Group words by first letter
words = ["Apple", "Artist", "Ball", "Bat", "Cartoon", "Car"]
groups = {}

for word in words:
  first = word[0]

  if first in groups:
    groups[first].append(word)
  else:
    groups[first] = [word]

print(groups)
```

```python
words = ["Apple", "Artist", "Ball", "Bat", "Cartoon", "Car"]
groups = {}

for word in words:
  groups.setdefault(word[0], []).append(word)

print(groups)
```

```python
words = ["Apple", "cartoon", "Ball", "Bat", "artist", "Car"]
groups = {}
for word in words:
  first = word[0]
  if first in groups:
    groups[first].append(word)
  else:
    groups[first] = [word]

print(groups)
```

```python
# 84. Merge two dictionaries
dict1 = {'A': 1, 'B': 2}
dict2 = {'C': 3, 'D': 4}
dict1.update(dict2)
print("Merged dictionary:", dict1)
```

```python
dic1 = {'A': 1, 'B': 2}
dic2 = {'C': 3, 'D': 4}
merged = {**dic1, **dic2}
print("Merged dictionary:", merged)
```

```python
dic1 = {'A': 1, 'B': 2}
dic2 = {'C': 3, 'D': 4}

result = dic1 | dic2
print("Merged dictionary:", result)
```

```python
# 85.	Sort dictionary by key
my_dict = {
    'c': 3,
    'a': 1,
    'b': 2
}

sorted_dict = dict(sorted(my_dict.items()))
print("Sorted dictionary by key:", sorted_dict)
```

```python
my_dict = {
    'c': 3,
    'a': 1,
    'b': 2
}

for key in sorted(my_dict):
    print(key,":",  my_dict[key])
```

```python
# 86.	Sort dictionary by value
my_dict = {
    'c': 3,
    'a': 1,
    'b': 2
}
sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))
print("Sorted dictionary by value:", sorted_dict)
```

```python
my_dict = {
    'a': 3,
    'b': 1,
    'c': 2
}
sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))
print("Sorted dictionary in descending order:", sorted_dict)
```

```python
# 87.	Remove duplicate elements using set
a = [1,2,3,4,2,3,5,1,2,1,2]
unique = list(set(a))
print("original list: ", a)
print("Unique elements after removing duplicates:", unique)
```

```python
# 88.	Find intersection of two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
intersection = set1.intersection(set2)
print("Intersection of sets:", intersection)
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7, 8}
result = set1 & set2
print("Intersection of sets", result)
```

```python
# 89.	Find union of two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
union = set1.union(set2)
print("Union of sets:", union)
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
result = set1 | set2
print("Union", result)
```

```python
# 90.	Check if two lists have common element
list1 = [1, 2, 3, 4, 5, 10]
list2 = [5, 6, 7, 8, 9, 10]
common_elements = set(list1) & set(list2)
print("Common elements:", common_elements)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
common_elements = set(list1) & set(list2)
print("Common elements:", common_elements)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
common_elements = False
for i in list1:
    if i in list2:
        common_elements = True
        break

print("Common elements:", common_elements)
```
